```python
import pandas as pd
import numpy as np
```

```python
#importing and merging data sets

school_data_to_import = "PyCitySchools\Resources\schools_complete.csv"
student_data_to_import = "PyCitySchools\Resources\students_complete.csv"

school_data = pd.read_csv(school_data_to_import)
student_data = pd.read_csv(student_data_to_import)

school_data_merge = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_merge.head()
```

```python
school_data.head()
```

```python
student_data.head()
```

```python
#calculating totals and means of schools and students

all_schools=school_data["school_name"].count()
all_schools

all_students=student_data["student_name"].count()
all_students

all_budget=school_data["budget"].sum()
all_budget

average_math=student_data[("math_score")].mean()
average_math


average_reading=student_data[("reading_score")].mean()
average_reading

average_both=((average_math + average_reading) /2)
average_both

#passing score >70
passing_math=(school_data_merge["math_score"]>70).sum() / school_data_merge["math_score"].count()* 100
passing_math 

#passing score reading >70
passing_reading= (school_data_merge["reading_score"]>70).sum() / school_data_merge["reading_score"].count()* 100
passing_reading
```

```python
#district summary data frame
schools_summary = pd.DataFrame({"Total Schools":[all_schools], "Total Students":[all_students],"Total Budget":[all_budget],
                                "Average Math Score":[average_math], "Average Reading Score":[average_reading], "% Passing Math":[passing_math],
                                "% Passing Reading":[passing_reading],"% Overall Passing Rate":[average_both]})
schools_summary
```

```python
#top Performing Schools
```

```python
#distinguish between charter and district
school_type=school_data.set_index(["school_name"])["type"]
school_type.head()
```

```python
#total number of students in each school
student_per_school=school_data_merge["school_name"].value_counts()
student_per_school.head()
```

```python
#school budget per school
school_budget=school_data_merge.groupby(["school_name"])["budget"].mean()
school_budget.head()
```

```python
#budget per student for each school
budget_per_student=school_budget/student_per_school
budget_per_student
```

```python
#mean math score
math_score_per_school=school_data_merge.groupby(["school_name"])["math_score"].mean()
math_score_per_school.head()
```

```python
#mean reading score
reading_score_per_school= school_data_merge.groupby(["school_name"])["reading_score"].mean()
reading_score_per_school.head()
```

```python
#school passing math score
pass_math_score=school_data_merge.loc[school_data_merge["math_score"]>=70]
pass_math_score

#school score assigned values.
math_score_per_school=pass_math_score["school_name"].value_counts()
math_score_per_school.head()

math_mean=math_score_per_school/student_per_school*100
math_mean
```

```python
#reading passing score
reading_score=school_data_merge.loc[school_data_merge["reading_score"]>=70]

#reading passing score assigned values
reading_score_per_school= reading_score["school_name"].value_counts()
reading_score_per_school.head()


reading_mean=reading_score_per_school/student_per_school*100
reading_mean
```

```python
#average of reading math passing scores per school
overall=math_mean + reading_mean/student_per_school
overall
```

```python
#new but old data frame for the school summary
sch_summary_df = pd.DataFrame({"School Type":school_type, "Total Students":student_per_school, "Total School Budget":school_budget
                                ,"Per Student Budget":budget_per_student ,"Average Math Score" :math_score_per_school,
                             "Average Reading Score":reading_score_per_school,
                             "% Passing Math":math_mean, "% Passing Reading":reading_mean,
                            "% Overall Passing Rate":overall})
sch_summary_df.head()
```

```python
#top performing schools
```

```python

top_school_df=sch_summary_df.sort_values(["% Passing Math","% Passing Reading","% Overall Passing Rate" ], 
                                        ascending=False)
top_school_df.head()
```

```python
#head for the top schools
top_school_df["Total School Budget"]=top_school_df["Total School Budget"].map("${:,.2f}".format)
top_school_df["Per Student Budget"]=top_school_df["Per Student Budget"].map("${:,.2f}".format)

top_school_df
```
